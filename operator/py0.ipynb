```python
#Print the last digit of the number
num = 5783
last_digit = num%10


print(last_digit)
```

```python
# Print the last two digit of the number
num = 5783
last_digit = num%100
print(last_digit)
```

```python
# Remove the last digit of the number
num = 5783
last_digit = num // 10
print(last_digit)
```

```python
# Remove the last two digit of the number
num = 5783
last_digit = num // 100
print(last_digit)
```

```python
# Print each digit of the number one by one from Right to Left.

num = 5783

while num > 0:
    rem = num % 10
    num = num // 10
    print(rem)
```
